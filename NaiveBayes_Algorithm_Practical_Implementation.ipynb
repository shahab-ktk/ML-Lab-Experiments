# **Naive Baye's Algorithm Practical Implementation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
X,y = load_iris(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

# **Training a model on Tips dataset by using Naive Baye's Algorithm**

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

```python
df['sex'].value_counts()
```

```python
df['smoker'].value_counts()
```

```python
df['day'].value_counts()
```

```python
df['time'].value_counts()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le1= LabelEncoder()
le2= LabelEncoder()
le3= LabelEncoder()
df['sex'] = le1.fit_transform(df['sex'])
df['smoker'] = le2.fit_transform(df['smoker'])
df['time'] = le3.fit_transform(df['time'])

```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('time',axis=1)
y = df['time']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=0)
```

```python
X_train.head()
```

```python
# One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_Cols = ['day']
ct = ColumnTransformer(
    transformers = [('encoder',OneHotEncoder(drop='first',handle_unknown='ignore'),categorical_Cols)],
    remainder = 'passthrough'
)
```

```python
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred = gnb.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
