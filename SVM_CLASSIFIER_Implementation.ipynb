```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000,n_classes=2,n_redundant=0,n_clusters_per_class=2,n_features=2)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0] ,y=pd.DataFrame(X)[1],hue = y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
# Prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train,y_train)
y_pred1 = rbf.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

```python
polynomial = SVC(kernel='poly')
polynomial.fit(X_train,y_train)
y_pred3 = polynomial.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
y_pred4 = sigmoid.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred4))
print(classification_report(y_test,y_pred4))
```

## Hyperparameter Tunning with SVC

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,cv = 5,verbose=3,n_jobs=-1)
grid.fit(X_train,y_train)
```

```python
y_pred5 = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred5))
print(classification_report(y_test,y_pred5))
```

## **SVM kernel Indepth Intuition and Practical Implementation**

```python
x = np.linspace(-5.0,5.0,100)
y = np.sqrt(10**2 - x**2)
y = np.hstack([y,-y])
x = np.hstack([x,-x])
```

```python
x1 = np.linspace(-5.0,5.0,100)
y1 = np.sqrt(5**2 - x1**2)
y1 = np.hstack([y1,-y1])
x1 = np.hstack([x1,-x1])
```

```python
plt.scatter(y,x,color='red')
plt.scatter(y1,x1,color='blue')
plt.show()
```

```python
df1 = pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y'] = 0
df2 = pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y'] = 1
df = pd.concat([df1, df2])
df.head()
```

```python
df.tail()
```

```python
# Independent and Dependent features
X = df.iloc[:,:2]
y = df.Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=10)
```

```python
# we need to find the components for the Polynomial kernels
# x1,x2,x1_square,x2_square,x1*x2
df["X1_Square"] = df['X1']**2
df["X2_Square"] = df['X2']**2
df["X1*X2"] = (df["X1"]*df["X2"])
df.head()
```

```python
# Independend and Dependent features
X = df[['X1','X2','X1_Square','X2_Square','X1*X2']]
y = df['Y']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=0)
```

```python
X_train
```

```python
import plotly.express as px
figure = px.scatter_3d(df,x = 'X1',y = 'X2',z = 'X1*X2',color='Y')
figure.show()
```

```python
import plotly.express as px
figure = px.scatter_3d(df,x = 'X1_Square',y = 'X2_Square',z = 'X1*X2',color='Y')
figure.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
classifier = SVC(kernel = 'linear')
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'poly')
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print(accuracy_score(y_test,y_pred))
```
