## **K Nearest Neighbour Classifier**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
X,y = make_classification(
    n_samples=1000,
    n_classes=2,
    n_features=3,
    n_redundant=1,
    random_state=999
    )
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5,algorithm='auto')
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# Applying Grid Search CV
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors':np.arange(1,50)}
grid = GridSearchCV(estimator=classifier,param_grid=param_grid,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
```

```python
y_pred1 = grid.predict(X_test)
```

```python
grid.best_params_
```

```python
print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

## **K Nearest Neighbour Regressor**

```python
from sklearn.datasets import make_regression
X,y = make_regression(
    n_samples=1000,
    n_features=2,
    noise=10,
    random_state=42
)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors=5,algorithm='auto')
regressor.fit(X_train,y_train)
```

```python
y_pred2 = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
print(r2_score(y_test,y_pred2))
print(mean_absolute_error(y_test,y_pred2))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors':np.arange(1,50)}
grid = GridSearchCV(estimator=regressor,param_grid=param_grid,cv=5,n_jobs=-1)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred3 = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
print(r2_score(y_test,y_pred3))
print(mean_absolute_error(y_test,y_pred3))
```
